## Use case
The use case for this notebook is to:
1. Capture images from the Arduino 33 BLE.
2. Preview the image.
3. If the image is of interest, it can be easily saved as ".png" and ".npy" format.

The sketch "Image_Dump_From_Arduino.ino" should be uploaded to the Arduino.

```python
from capture_img import read_img, save_numpy, save_img, img_from_565
import serial
import numpy as np
from PIL import Image
from matplotlib import pyplot as plt

port = "/dev/ttyACM0"
baud = 9600
verbose = False
#width = 320
#height = 240
width = 176
height = 144
bytes_per_pixel = 2
#file_prefix = "../../datasets/Images_for_Training/Maria/img_HighRes"
file_prefix = "../../datasets/Images_for_Training/PATH_HERE"

id = 0
```

```python
with serial.Serial(port, baud) as ser:
    ser.write(b'c') # send 'c' to start capture
    img_data = read_img(ser, width, height, bytes_per_pixel, verbose=verbose)

    plt.axis('off')
    plt.imshow(img_from_565(img_data))
```

```python
save_numpy(img_data, file_prefix + "_" + str(id) + ".npy")
save_img(img_data, file_prefix + "_" + str(id) + ".png")
id += 1
```
